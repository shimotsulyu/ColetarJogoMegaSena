```python
def gerar_tabela(arquivo,data=None,chromedriver='/usr/bin/chromedriver'):
    from selenium import webdriver
    from selenium.webdriver.common.keys import Keys
    driver = webdriver.Chrome(chromedriver)
    driver.get('file://'+arquivo)
    time.sleep(1)
    col = ['Concurso', 'Data_Sorteio', '1Dezena', '2Dezena', '3Dezena', '4Dezena', '5Dezena', '6Dezena']
    concursos = len(driver.find_elements_by_xpath('//tr/td[1]'))
    if data is None:
        print('Gerando Tabela')
        data = pd.DataFrame(columns=col)
        for k in range(concursos):
            sorteio = driver.find_element_by_xpath('//tr['+str(k+2)+']').text.split()[0:8]
            if len(sorteio)==8:
                data = data.append(pd.DataFrame([sorteio],columns=col), ignore_index=True)
    else:
        print('Último concurso:',driver.find_element_by_xpath('//tr['+str(concursos)+']').text.split()[1])
        if data['Concurso'].tail(1).to_string(index=False)[1:] == driver.find_element_by_xpath('//tr['+str(concursos)+']').text.split()[0]:
            print('Já está Atualizado')
        else:
            print('Atualizando Tabela')
            sorteio = driver.find_element_by_xpath('//tr['+str(concursos)+']').text.split()[0:8]
            data = data.append(pd.DataFrame([sorteio],columns=col), ignore_index=True)
    return data
def importar_jogos(url):
    import requests, zipfile, io 
    r = requests.get(url)
    z = zipfile.ZipFile(io.BytesIO(r.content))
    z.extractall()
```

```python
import glob
import os
import time
import pandas as pd
url = 'http://www1.caixa.gov.br/loterias/_arquivos/loterias/D_megase.zip'
importar_jogos(url)
arquivo = glob.glob('*.htm')
ini = time.time()
data = gerar_tabela(os.path.abspath(arquivo[0]))
fim = time.time()
print(fim-ini,'s')
data.to_csv('megase.csv')
data
```
